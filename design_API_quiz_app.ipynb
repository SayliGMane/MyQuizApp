Register the user

- `/api/v1/user/registration/`
    - `POST` : Register the user
        - **Header**
            - Content-Type: `application/json`
            - Accept: `application/json`
        - **Request body**
            - **Fields**
                - username (str)
                    - required `true`, unique `true`
                - first_name (str)
                    - required `false`, default `''`
                - last_name (str)
                    - required `true`
                - password (str)
                    - required `true`
                - email (str)
                    - required `true`, unique `true`
        
        - **Example**
    
            ```json
                {
                    "username": "kevin",
                    "first_name": "kenz",
                    "last_name": "eyong",
                    "password": "*********",
                    "email": "tony@gmail.com",
                }
            ```

        - Request Response
            - Status code: `201 Created`

            ```json
                {
                    "message": "Successfully registrated. Welcome to Quiz App",
                    "username": "kevin"
                }
            ```

        - Request Response
            - Status code: `400 Invalid Input`

            ```json
                {
                    "errorMessage": "Something went wrong please try again."
                }
            ```
